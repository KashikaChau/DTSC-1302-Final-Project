```python
# dependencides 
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
warnings.filterwarnings("ignore")   # Suppress all warnings
```

```python
# dataset, only with data of interest
df = pd.read_csv(r"C:\Users\Adele\Downloads\data of interest - DPIC Execution Database - U.S. Executions (1) (1).csv")

#converting 'yes' and 'no' into 1 and 0
df['Volunteered'] = df['Execution Volunteer'].map({'no': 0, 'yes': 1})
df.head(9)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# boxplot
sns.boxplot(x='Volunteered', y='Number of Victims', data=df)
plt.show()
```

```python
# violin plot
sns.violinplot(x='Volunteered', y='Number of Victims', data=df)
plt.show()
```

```python
# bar plot with median
median_values = df.groupby('Volunteered')['Number of Victims'].median()
median_values.plot(kind='bar')
plt.show()
```

```python
# bar plot with mean
sns.barplot(x='Volunteered', y='Number of Victims', data=df, estimator='mean')  # For mean
plt.show()
```

```python
# histogram
sns.histplot(df[df['Volunteered'] == 0]['Number of Victims'], color='blue', kde=True, label='Category 0')
sns.histplot(df[df['Volunteered'] == 1]['Number of Victims'], color='orange', kde=True, label='Category 1')
plt.legend()
plt.show()
```

```python
# scatter plot with jitter
sns.stripplot(x='Volunteered', y='Number of Victims', data=df, jitter=True)
plt.show()
```

```python
# swarm plot
sns.swarmplot(x='Volunteered', y='Number of Victims', data=df)
plt.show()
```

```python
# point plot
sns.pointplot(x='Volunteered', y='Number of Victims', data=df, estimator='mean')
plt.show()
```
